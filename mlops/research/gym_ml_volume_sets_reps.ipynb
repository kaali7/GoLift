## **Model**

```python
model_Df = clean_Df.groupby('exercise_id').agg({
    "muscle_group":"first",
    "body_part":"first",
    "phase":"first",
    "difficulty_level":"first",
    "equipment":"first",
    "day_type":"first",
    "sets_planned":"mean",
    "reps_planned":"mean",
    "weight_planned":"mean",
    "sets_completed":"mean",
    "reps_completed":"mean",
    "weight_used":"mean",
    "rpe":"mean",
    "rest_time_sec":"mean",
    "completion_rate":"mean",
    "next_sets":"mean",
    "next_reps":"mean",
    "next_weight":"mean",
}).reset_index()

model_Df.head()
```

```python
model_Df.columns
```

### **Sets Model**

```python
sets_Df = model_Df[['exercise_id', 'muscle_group', 'body_part', 'phase', 'difficulty_level','equipment', 
                    'day_type', 'sets_planned','sets_completed','rest_time_sec', 'completion_rate', 'next_sets']]

sets_Df.head()
       
```

```python
set_x = sets_Df.drop(columns='next_sets')
set_y = sets_Df[['next_sets']]
```

```python
# here we are using one hot coding for encode the categorical columns
set_object_cols = set_x.select_dtypes("object").columns
set_num_cols = set_x.select_dtypes(include=['int64', 'float64']).columns
```

```python
set_preprocessor= ColumnTransformer(
    transformers=[
        ('object', OneHotEncoder(sparse_output=False, drop='first'), set_num_cols),
        ('numerical', StandardScaler(), set_num_cols)
    ]
)

sets_pipeline = Pipeline(steps=[
    ('preprocessor', set_preprocessor),
    ('regression_model', LinearRegression())
])
```

```python
sets_pipeline.fit(set_x, set_y)
```

```python
X_train, X_test, y_train, y_test = train_test_split(set_x, set_y, test_size=0.2, random_state=42)
```

```python
# Evaluate
score = sets_pipeline.score(X_test, y_test)
print(f"Model R² score: {score:.3f}")

# Make predictions
predictions = sets_pipeline.predict(X_test)
print(f"Predictions: {predictions[:5]}")  # First 5 predictions
```
